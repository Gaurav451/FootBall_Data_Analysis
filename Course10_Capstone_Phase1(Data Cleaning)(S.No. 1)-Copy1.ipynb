```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# try appearance and games merging for future improvements
```

```python
df1 = pd.read_excel('players.xlsx')
df1.head()
```

```python
df2 = pd.read_excel('games.xlsx')
df2.head()
```

```python
df3 = pd.read_excel('game_lineups.xlsx')
df3.head()
```

```python
df4 = pd.read_excel('game_events.xlsx')
df4.head()
```

```python
df5 = pd.read_excel('appearances.xlsx')
df5.head()
```

```python
# df2,df3
df_concat_1 = df2.merge(df3, on = 'game_id', how = 'outer')
df_concat_1
```

```python
# df5,df1
df_concat_2 = df1.merge(df5, on = 'player_id', how = 'outer') 

df_concat_2
```

```python
df_concat_2.isnull().sum()
```

```python
df_concat_3 = df_concat_2.merge(df2, on = 'game_id', how = 'outer')
df_concat_3.head()
```

```python
df_concat_4 = df_concat_3.merge(df3, on = 'game_id', how = 'outer')
df_concat_4.head()
```

```python
df_concat_4.columns
```

```python
df_concat_5 = df_concat_4.merge(df4, on = 'game_id', how = 'outer')
df_concat_5.head()
```

```python
# pd.set_option('display.max_columns', None)
df_concat_5.iloc[19]
```

```python
df_concat_5.isnull().sum()
```

```python
# We will drop the columns which are showing a high percentage of null values
df_concat_5 = df_concat_5.drop(['home_club_formation','away_club_formation','game_lineups_id','number','player_id_y','player_name_y','team_captain','position_y','player_assist_id','description','player_in_id'],axis = 1)
df_concat_5.head()
```

```python
df_concat_5.isnull().sum()
```

```python
# removing date_y,competition_id_y column from date_x and date_y, competition_id_x and competition_id_y columns, as both are one and same only
# player_id column will also be dropped as it is a redundant copy of column 'player_id_x' and it has lots of null values also

df_concat_5 = df_concat_5.drop(['date_y','competition_id_y','player_id'],axis = 1)

df_concat_5.head()
```

```python
# date column must also be removed as it is a redundant copy of date_x and contains many null values
df_concat_5 = df_concat_5.drop(['date'],axis = 1)
df_concat_5.head()
```

```python
df7 = df_concat_5.select_dtypes(exclude = 'number')
df7.head()
```

```python
df7.isnull().sum()
```

```python
cols = ['contract_expiration_date','agent_name','type_y','type_x']

for i in cols:
    df_concat_5[i].fillna('Unknown',inplace = True)

df_concat_5.isnull().sum()
```

```python
for i in df7.columns:
    df_concat_5[i].fillna(df_concat_5[i].mode()[0],inplace = True)

df_concat_5.isnull().sum()
```

```python
df_concat_5 = df_concat_5.drop(['game_event_id'],axis = 1)
```

```python
df_concat_5.isnull().sum()
```

```python
num_cols = df_concat_5.select_dtypes(include = 'number')

df_concat_5[num_cols.columns] = df_concat_5[num_cols.columns].apply(lambda x : x.interpolate())

df_concat_5.isnull().sum()
```

```python
df_concat_5.info()
```

```python
df_concat_5['minute'].fillna(0.0,inplace = True)
```

```python
df_concat_5['home_club_position'].fillna(0.0,inplace = True)
df_concat_5['away_club_position'].fillna(0.0,inplace = True)
```

```python
df_concat_5.info()
```

```python
# Filtering the dataset for the condition below, as both the columns mean the same, and then deleting the player_name_x column
```

```python
df_concat_5 = df_concat_5[df_concat_5['name'] == df_concat_5['player_name_x']]
df_concat_5.shape
```

```python
#df_concat_5 = df_concat_5.drop(['player_name_x'],axis = 1)
df_concat_5.info()
```

```python
df_concat_5.columns = df_concat_5.columns.str.replace('_x','')
```
